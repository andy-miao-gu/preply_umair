```python
import pandas as pd
data = pd.read_csv('StudentsPerformance.csv')
```

```python
data
```

```python
import seaborn as saor
import matplotlib.pyplot as mtltippo
mtltippo.figure(figsize=(10, 4))
saor.violinplot(data=data, x="parental level of education", y="math score",split=True,hue ='gender')
```

```python
mtltippo.figure(figsize=(10, 4))
saor.violinplot(data=data, x="parental level of education", y="math score",inner="q",split=True, bw_adjust=0.2 )
```

```python
mtltippo.figure(figsize=(10, 4))
saor.violinplot(data=data, x="parental level of education", y="math score",inner='stick')
```

```python
mtltippo.figure(figsize=(10, 4))
saor.violinplot(data=data, x="parental level of education", y="math score",bw=0.999999999999999)
```

```python
mtltippo.figure(figsize=(10, 4))
saor.swarmplot(data=data, x="parental level of education", y="math score",hue ='gender')
```

```python
mtltippo.figure(figsize=(10, 4))
saor.boxenplot(data=data, x="parental level of education", y="math score",hue ='gender')
```
