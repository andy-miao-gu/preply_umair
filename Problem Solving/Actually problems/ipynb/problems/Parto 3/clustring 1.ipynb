# prepare Data

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
data, labels =  make_blobs(n_samples=300, n_features=2, centers=3)

# plot and confirm
plt.scatter(data[:,0],data[:,1])
K=3
```

# making model

```python
# import numpy and initialize k = 3 max_it = 10 and tol = 1e-4
import numpy as np


k = 3
max_it = 10
tol = 1e-4

# intialize number centroids using random function

centroids = data[ np.random.choice( data.shape[0],K ,replace=False)]
centroids 
```

```python
# where are our centrolds right now?
# plot and confirm
plt.scatter(data[:,0],data[:,1])
plt.scatter(centroids[:,0],centroids[:,1])
```

```python
iteration = 0
labels = np.argmin(np.linalg.norm(data[:, np.newaxis] - centroids, axis=2), axis=1)
labels
# Visualization of clusters (optional, just for demonstration)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')
plt.title(f"Iteration {iteration + 1}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have 'data' and 'centroids' defined
K = len(centroids)

for i in range(10):
    labels = np.argmin(np.linalg.norm(data[:, np.newaxis] - centroids, axis=2), axis=1)

    # Visualization of clusters (optional, just for demonstration)
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')
    plt.title(f"Iteration {i + 1}")
    plt.show()

    centroids_n = np.array([data[labels == K].mean(axis=0) if np.sum(labels == K) > 0 else data[np.random.choice(len(data))] for K in range(K)])

    # Step 4: Convergence Check
    if np.all(np.abs(centroids_n - centroids) < 1e-5):
        break

    centroids = centroids_n.copy()
```
