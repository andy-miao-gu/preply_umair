using open(file.path) load file and change ti into nested dictionary

and at the end convert it into data frame and 


plot fruits vs price as bar plot

```python
x = open("mango.txt", "r")
data = x.readlines()
data
```

```python
listo=[]
```

```python
data[0].split('\\t')
```

```python
for each in data:
    listo.append(each.split("\\t"))
```

```python
listo
```

```python
lines = []
words = []

for line in listo:
    for word in line:
        words.append(word.split("\\n")[0])
    lines.append(words)
    words = []
print(lines)
```

```python
lines
```

```python
d = {}
dd = {}
i = 0
#### home workd 
for lins in lines:
    if lins[0] != 'Fruit Name':
        dd['name'] = lins[0]
        dd['price']  = float(lins[1])
        dd['color'] = lins[2]
        dd['shape'] = lins[3]
    else:
        continue
    d[i] = dd
    dd = {}
    i += 1
print(d)
## make a dictionary
```

```python
import pandas as pd
bruh = pd.DataFrame(d).transpose()
bruh['price'] = bruh['price'].astype(float)
```

```python
import seaborn as see
import matplotlib.pyplot as plt
plt.figure(figsize=(25,20),dpi=300)
see.barplot(x="name",y="price",data=bruh)
```
