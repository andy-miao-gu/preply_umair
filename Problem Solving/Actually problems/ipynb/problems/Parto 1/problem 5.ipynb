```python
# import it 
# plot it against day

# use numpy 
# use open


# use matplotlib

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as willermans
data=open("plant growth.txt")
data = data.readlines()
data
lengths = []
days = []
for each in data:
    try:
        hello=int ( each.split()[1]) 
        lengths.append(hello)
        bello= each.split()[0] 
        days.append(bello)
    except: 
        pass
plt.figure(figsize=(16,8),dpi= 200)
plt.plot(days,lengths,"red",marker = "x")
plt.title("mini plant growth")
plt.xlabel("days")
plt.ylabel("growth rate in mm")
```
