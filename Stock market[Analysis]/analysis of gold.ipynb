```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
gold = yf.Ticker("GC=F")

start_date = "2020-01-01"
end_date = "2023-01-01"
```

```python
gold_data =  gold.history(period="id",start=start_date,end = end_date   )
```

```python
gold_data
```

```python
plt.figure(dpi=675)
sns.lineplot(gold_data[['Open','High']])
```

```python
gold_data = gold_data.reset_index()
```

```python
gold_data.columns
```

```python
gold_data.sort_values('Open',ascending=False)
```

```python
plt.figure(figsize=(16,8),dpi=675)
import plotly.express as px

# Assuming 'Date' and 'Volume' are columns in your DataFrame
fig = px.scatter(gold_data, x='Date', y='Volume', title='Scatter Plot of Gold Data')
fig.show()
```
