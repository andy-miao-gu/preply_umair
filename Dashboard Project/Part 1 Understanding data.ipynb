```python

import chardet

# Step 1: Detect the file's encoding
file_path = 'Global YouTube Statistics.csv'

with open(file_path, 'rb') as file:
    result = chardet.detect(file.read())
    detected_encoding = result['encoding']

print(f"Detected Encoding: {detected_encoding}")
```

```python
import plotly.graph_objects as go
from dash import Dash, dcc, html

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Sales and Profit Dashboard"),
    dcc.Graph(
        id='line-plot',
        figure={
            'sub_data': [
                go.Scatter(x=sub_data['Youtuber'], y=sub_data['subscribers'], mode='lines', name='subscribers'),
                go.Scatter(x=sub_data['Youtuber'], y=sub_data['video views'], mode='lines', name='video views'),
            ],
            'layout': {
                'title': 'Sales and Profit Over Time',
                'xaxis': {'title': 'Date'},
                'yaxis': {'title': 'Amount'}
            }
        }
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

# example of Dasshboard

```python
# Example data (you can load your own dataset)
data = {
    'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04'],
    'Sales': [100, 120, 90, 110],
    'Profit': [20, 25, 18, 22]
}

df = pd.DataFrame(data)
```

```python
import plotly.graph_objects as go
from dash import Dash, dcc, html

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Sales and Profit Dashboard"),
    dcc.Graph(
        id='line-plot',
        figure={
            'data': [
                go.Scatter(x=df['Date'], y=df['Sales'], mode='lines', name='Sales'),
                go.Scatter(x=df['Date'], y=df['Profit'], mode='lines', name='Profit')
            ],
            'layout': {
                'title': 'Sales and Profit Over Time',
                'xaxis': {'title': 'Date'},
                'yaxis': {'title': 'Amount'}
            }
        }
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
