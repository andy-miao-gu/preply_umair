```python
def moneycalcu(cuisine00,cuisine01):
    cuisine1 = data [ (data['cuisine_type']==cuisine00) | (data['cuisine_type']==cuisine01)]
    numpy.boxplot(x=cuisine1['cost_of_the_order_yen'])
    plt.show()
    cuisine2 = data [ (data['cuisine_type']==cuisine00) | (data['cuisine_type']==cuisine01)]
    numpy.boxplot(x=cuisine2['cost_of_the_order_yuan'])
    plt.show()
    cuisine3 = data [ (data['cuisine_type']==cuisine00) | (data['cuisine_type']==cuisine01)]
    numpy.boxplot(x=cuisine3['cost_of_the_order'])
    plt.show()
    cuisine4 = data [ (data['cuisine_type']==cuisine00) | (data['cuisine_type']==cuisine01)]
    numpy.boxplot(x=cuisine4['cost_of_the_order_pakirupe'])
    plt.show()
```

```python
def moneycalcu(df):
   numpy.boxplot(x="cost_of_the_order",y="cuisine_type",data=df)
   plt.show()
   numpy.boxplot(x="cost_of_the_order_pakirupe",y="cuisine_type",data=df)
   plt.show()
```
