```python
import pandas as pd
import numpy as np

# Creating a DataFrame with random values
np.random.seed(42)  # For reproducibility

data = {
    'Feature1': np.random.randint(1, 50, size=30),
    'Feature2': np.random.randint(1, 50, size=30)
}

df = pd.DataFrame(data)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math as mt
from sklearn.cluster import KMeans
import matplotlib.image as mpimg
```

```python
sns.scatterplot(x="Feature1",y="Feature2",data=df)
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=3)  # Specify the number of clusters
kmeans.fit(df)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Add the cluster labels to the DataFrame
df['Cluster'] = cluster_labels
cluster_centers = kmeans.cluster_centers_
# Print the DataFrame with cluster labels
print(df)
sns.scatterplot(x="Feature1",y="Feature2",data=df,style="Cluster")
plt.show()
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='red', s=200, label='Cluster Centers')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Scatter Plot with Cluster Boundaries')
plt.legend()
plt.show()
```
