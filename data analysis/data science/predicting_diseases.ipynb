```python
import pandas as adsisumair
```

```python
df = adsisumair.read_csv("gene_expression.csv")
```

```python
import seaborn as ebrisumair
df.columns = ['Gene One','Gene Two','ILL or not']
```

```python
ebrisumair.scatterplot(data=df, x="Gene One", y = "Gene Two", hue="ILL or not")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.drop('ILL or not',axis=1)
y = df['ILL or not']
```

```python
knnmodel = KNeighborsClassifier(n_neighbors=1)
```

```python
knnmodel.fit(X,y)
```

```python
if knnmodel.predict([[10,1]]):
    print("this patient is ill")
else:
    print("you are well")
```
