#### Fibonacci Sequence
Write a program to generate the Fibonacci sequence up to a given term n.

```python
import math as mt
import matplotlib.pyplot as plt
F0 = 0
F1 = 1
F2 = F1 + F0
F3 = F2 + F1 
F4 = F3 + F2
F5 = F4 + F3
print(F5)
#F10??
F6 = F5 + F4
F7 = F6 + F5
F8 = F7 + F6
F9 = F8 + F7
F10 = F9 + F8
print(F10)
```

```python

def function(parameter):
    F0 = 0
    F1 = 1
    Fn = 0
    for i in range(2, parameter + 1):
        Fn = F0 + F1
        F0 = F1 
        F1 = Fn
    return Fn


# make a list of 1 to 20
inp = [each for each in range(2,1000) ]
# fnd the fibnoshi of each number 
out = [function(each) for each in inp]

# plot it 
plt.plot(inp, out)

```

#### Matrix Operations
Develop a program for matrix multiplication, addition, and subtraction.

```python
def matrix(mys,mys2):
    mys = np.array(mys)
    mys2 = np.array(mys2)
    x=mys+mys2
    y=mys-mys2
    z=mys*mys2
    return {"sum":x,"diff":y,"prd":z}
matrix([[1,2],[3,4]],[[5,6],[8,9]])
```
